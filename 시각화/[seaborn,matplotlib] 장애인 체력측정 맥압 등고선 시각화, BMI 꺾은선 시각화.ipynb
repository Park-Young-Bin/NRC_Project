```python
import warnings
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

```python
iris= sns.load_dataset('iris')
iris.info()
```

```python
sns.kdeplot(iris.sepal_width, iris.sepal_length, cmap="Reds", shade=True, bw=.15)
```

```python
sns.kdeplot(iris.sepal_width, iris.sepal_length, cmap="Blues", shade=True, shade_lowest=True)
```

```python
sns.kdeplot(iris.sepal_width, iris.sepal_length, n_levels=30, cmap="Purples_d")
```

```python
setosa = iris.loc[iris.species == "setosa"]

virginica = iris.loc[iris.species == "virginica"]

ax = sns.kdeplot(setosa.sepal_width, setosa.sepal_length,

cmap="Reds", shade=True, shade_lowest=False)

ax = sns.kdeplot(virginica.sepal_width, virginica.sepal_length,

cmap="Blues", shade=True, shade_lowest=False)
```

```python
df=pd.read_excel("장애인 체력 측정 항목별 측정 데이터(2021.01.~2021.08.).xlsx")
df.head()
```

```python
df=df.fillna("")
```

```python
df.info()
```

```python
지적장애 = df.loc[df.장애유형명 == "지적장애"]
시각장애 = df.loc[df.장애유형명 == "시각장애"]
청각장애 = df.loc[df.장애유형명 == "청각장애"]
척수장애 = df.loc[df.장애유형명 == "척수장애"]
```

```python
plt.rc('font', family='NanumGothic') #한글폰트
```

```python
figure, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2)
figure.set_size_inches(16, 10)
ax1 = sns.kdeplot(지적장애.측정연령수, 지적장애.맥압,cmap="Blues", shade=True,thresh=0.05,ax=ax1)# shade_lowest=False)
ax2 = sns.kdeplot(시각장애.측정연령수, 시각장애.맥압,cmap="Oranges", shade=True, shade_lowest=False,ax=ax2)
ax3 = sns.kdeplot(청각장애.측정연령수, 청각장애.맥압,cmap="Purples", shade=True, shade_lowest=False,ax=ax3)
ax4 = sns.kdeplot(척수장애.측정연령수, 척수장애.맥압,cmap="Greens", shade=True, shade_lowest=False,ax=ax4)

ax1.set(title="지적장애")
ax2.set(title="시각장애")
ax3.set(title="청각장애")
ax4.set(title="척수장애")
```

```python
fig,(ax1,ax2,ax3,ax4)= plt.subplots(nrows=4)
fig.set_size_inches(18,25)

sns.pointplot(data=지적장애, x="측정연령수", y="05 :BMI", ax=ax1)
sns.pointplot(data=시각장애, x="측정연령수", y="05 :BMI", ax=ax2)
sns.pointplot(data=청각장애, x="측정연령수", y="05 :BMI", ax=ax3)
sns.pointplot(data=척수장애, x="측정연령수", y="05 :BMI", ax=ax4)
ax1.set(title="지적장애")
ax2.set(title="시각장애")
ax3.set(title="청각장애")
ax4.set(title="척수장애")
```
