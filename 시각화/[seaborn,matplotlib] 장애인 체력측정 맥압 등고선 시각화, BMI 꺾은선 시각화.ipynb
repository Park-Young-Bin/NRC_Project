```python
import warnings
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

```python
iris= sns.load_dataset('iris')
iris.info()
```

```python
sns.kdeplot(iris.sepal_width, iris.sepal_length, cmap="Reds", shade=True, bw=.15)
```

```python
sns.kdeplot(iris.sepal_width, iris.sepal_length, cmap="Blues", shade=True, shade_lowest=True)
```

```python
sns.kdeplot(iris.sepal_width, iris.sepal_length, n_levels=30, cmap="Purples_d")
```

```python
setosa = iris.loc[iris.species == "setosa"]

virginica = iris.loc[iris.species == "virginica"]

ax = sns.kdeplot(setosa.sepal_width, setosa.sepal_length,

cmap="Reds", shade=True, shade_lowest=False)

ax = sns.kdeplot(virginica.sepal_width, virginica.sepal_length,

cmap="Blues", shade=True, shade_lowest=False)
```
