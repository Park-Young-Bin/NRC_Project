```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['font.size'] = 11
```

```python
plt.figure(figsize = (6, 6))
ratio = [56.1, 15.9, 8.5,  7.3, 6.1, 6.1]
labels = ['지체장애', '뇌병변장애', '시각장애', '청각장애', '기타장애', '지적장애']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 2}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle = 90, counterclock=False, wedgeprops=wedgeprops)
plt.title('서울시 장애유형별 운영 동호회 현황', fontsize = 17)
plt.savefig('pyvisual/서울시 장애유형별 운영 동호회 현황.png', dpi=300)
plt.show()
```
