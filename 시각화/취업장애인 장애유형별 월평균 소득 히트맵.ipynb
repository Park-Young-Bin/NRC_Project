## 취업장애인 장애유형별 월평균 소득 히트맵

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("취업장애인 장애유형별 월평균 소득.xlsx",index_col=0)
df['다년평균']=df.mean(axis=1)
df
```

```python
plt.rc('font', family = 'NanumGothic')
```

```python
# 범죄 검거 비율 heatmap으로 시각화
target_col = [2000,2005, 2008, 2011, 2014,2017,'다년평균']

df_sort = df.sort_values(by='다년평균', ascending=False)
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(df_sort[target_col], annot=True, fmt='.0f',
           linewidths=.5, cmap='YlGnBu')
plt.title('취업장애인 장애유형별 월평균 소득 히트맵(다년평균순)',fontsize=15,y=1.03)
plt.legend(title="단위:만원",loc='upper right', bbox_to_anchor=(1.07, 1.07),fontsize=10)
plt.show()
```
