```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('pydata/장애인 체력 측정별 운동처방 데이터(2021.01._2021.08.).xlsx')
df.head()
```

```python
# 필요한 변수만 추출
df1 = df[['성별구분코드', '장애유형명', '연령대','장애등급명', '운동처방내용']]
df1.head()
```

```python
# 운동처방내용 ':' 기준으로 분리
result = df['운동처방내용'].str.split('/')
result = result.apply(lambda x: pd.Series(x))
result
```

```python
# df2(= df1 + result) 데잍터프레임 결합
df2 = pd.concat([df1.drop(columns = '운동처방내용'), result], axis=1)
df2
```

```python
df3 = pd.melt(df2, id_vars=['성별구분코드', '장애유형명', '연령대','장애등급명']).sort_values(by = ['성별구분코드', '장애유형명', '연령대','장애등급명'])
df3.drop(columns='variable', inplace=True)
```
