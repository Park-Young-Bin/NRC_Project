### step3. 데이터프레임 합치기(concat)

### step4. 결측값 및 문자열 -> 값으로 대체
- nan, 없음 -> 0
-  ○, 있음 -> 1

```python
# nan to 0
merge_df.fillna(value = 0, inplace = True)
merge_df.head()
```

```python
col_list = merge_df.columns.tolist()[2:]
```

```python
col_list
```

```python
for col in col_list:
    merge_df.replace({col : '있음'}, {col : 1}, inplace = True)
    merge_df.replace({col : '없음'}, {col : 0}, inplace = True)
    merge_df.replace({col : '○'}, {col : 1}, inplace = True)
    merge_df.replace({col : 'O'}, {col : 1}, inplace = True)
merge_df.head()
```

```python
merge_df.info()
```

```python
merge_df[['역명', '시군구 이름']]
```

### step5. 같은 역이름은 같은 위경도 값 넣기 / 편의시설 개수가 다른 것은 더 큰 값을 넣기

```python
melt_df = merge_df.drop(columns='index').melt(id_vars=['호선', '역명', '시군구 이름'])
melt_df.head()
```

```python
pivot_df = pd.pivot_table(melt_df, index = '역명', columns = 'variable', values = 'value', aggfunc='max')
# pivot_df = pivot_df.drop(columns='variable')
```

```python
pivot_df
```

```python
pivot_df.columns
```

### step6. 호선 데이터 + 역명 데이터 merge

```python
line_df = merge_df[['호선', '역명', '시군구 이름']]
```

```python
final_df = pd.merge(pivot_df, line_df, how='inner', on='역명').drop_duplicates().set_index('호선').sort_values(by=['호선', '역명'], ascending=[True, True]).reset_index()
final_df
```

```python
final_df = final_df.drop(columns='시군구 이름').drop_duplicates()
```

```python
final_df
```

```python
city_name = df2[['역명', '시군구 이름']]
city_name
```

### step7. final_df + city_name 데이터 merge

```python
final_df2 = pd.merge(city_name, final_df, how='inner', on='역명').drop_duplicates()
final_df2
```

### step8. 점자유도블럭 vs 점자유도타일 중 더 큰 값을 적용하여 '점자유도블럭' 컬럼에 삽입

```python
# 점자유도블럭, 점자유도타일 컬럼만 추출
a = final_df2[['점자유도블럭', '점자유도타일']]
a.head()
```

```python
# 크기 비교 후, 데이터프레임(b) 생성
b = a.loc[:, ['점자유도블럭', '점자유도타일']].fillna(0).max(axis=1)
b = pd.DataFrame(b)
b.columns = ['점자유도블럭']
b.head()
```

```python
# final_df2 데이터의 '점자유도블럭', '점자유도타일' 컬럼을 삭제하고 b와 데이터 셋 결합
final_df2.drop(columns=['점자유도블럭', '점자유도타일'], inplace = True)
final_df2.columns
```

```python
# 합친 데이터(final_df3) 생성
final_df3 = pd.concat([final_df2, b],axis=1)
final_df3.head()
```

```python
final_df3.to_excel('pydata/지하철 편의시설 현황 최종 취합_final.xlsx', index = False)
```
