```python
from scipy.stats import chi2_contingency
```

```python
df1=df[df['01)만성질환명(고혈압)']!=0]
고혈압=pd.crosstab(df1['가장 불편한부위2'],df1['01)만성질환명(고혈압)'],margins=True)
#교차분석 결과값들 일괄계산
#correction=True를 적용하면 Yates' correction이 적용 되어 검정통계량이 보수적으로 더 낮게 나옴
result고혈압=chi2_contingency(observed=고혈압, correction=False)
result고혈압
print("1. 카이제곱 통계량:", result고혈압[0])
print("2. p-value:", result고혈압[1])
print("3. df:", result고혈압[2]) #(행의개수-1)*(열의개수-1)
print("4. 기대값 행렬:")
pd.DataFrame(result고혈압[3]).rename(index={0:'절단', 1:'마비',2:'관절장애',3:'ALL'}, columns={ 0:'O',1:'X',2:'ALL'})
```
