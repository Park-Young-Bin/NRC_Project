### step4. 보조기기가 필요하지만 소지하지 않은 이유(H3)

```python
# H3 컬럼 추출
need_not_have = cp_df[['H3']]
need_not_have
```

```python
# H3값 빈도수 확인 후 데이터프레임 전환
needtohave = pd.DataFrame(need_not_have['H3'].value_counts())
needtohave
```

```python
# H3값별 비율 구하기(비해당 포함)
total = []

for i in range(len(pd.DataFrame(need_not_have['H3'].value_counts()))):
    total.append(42)

needtohave['total'] = total # 총 개수 합을 total 컬럼에 넣기

needtohave['pct'] = round(np.divide(needtohave.H3, needtohave.total) * 100, 2) # 비율 게산 (빈도수/총개수 * 100)

needtohave
```

```python
needtohave_1 = needtohave.iloc[1:, :]
needtohave_1
```

```python
# H3값별 비율 구하기(비해당 제거)

needtohave_1 = needtohave.iloc[1:, :]

total = []

for i in range(len(needtohave_1)):
    total.append(20)

needtohave_1['total'] = total # 총 개수 합을 total 컬럼에 넣기

needtohave_1['pct'] = round(np.divide(needtohave_1.H3, needtohave_1.total) * 100, 2) # 비율 게산 (빈도수/총개수 * 100)

needtohave_1
```

### step5. 파이차트 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family = 'Malgun Gothic')
```

```python
plt.figure(figsize = (8, 8))
ratio = needtohave_1['pct'].tolist()
labels = ['구입비용 부담', '본인에 적합한 보조기기 모름', '구입처 모름', '효과 없을 것 같음', '불편할 것 같음', '미관상 흉함', '적합한 보조기기 없음']
colors = ['#6e9fed', '#a0afff', '#b4c3ff', '#d2e1ff', '#dcebff', '#e8f5ff', '#f4ffff']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 2}

plt.pie(ratio, labels=labels, colors=colors, autopct='%.1f%%', startangle = 90, counterclock=False, wedgeprops=wedgeprops)
plt.title('장애인보조기기 필요하지만 소지하지 않은 이유', fontsize = 17)
theme = plt.get_cmap('bwr')
plt.savefig('pyvisual/장애인보조기기 필요하지만 소지하지 않은 이유.png', dpi=300)
# plt.legend()
plt.show()
```
