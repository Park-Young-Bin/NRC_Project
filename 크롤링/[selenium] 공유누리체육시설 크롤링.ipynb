```python
import urllib.request
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
import pandas as pd
```

```python
#요일별 이용시간 : tr 5개 
평일=[]
토요일=[]
일요일_공휴일=[]
접수일시=[]
이용시간=[]

#기본 정보
자원분류=[]
자원명칭=[]
장소_위치=[]
제공기관=[]
담당자정보=[]
예약방법=[]
예약문의=[]
이용대상=[]
준비시간=[]
신청서식=[]
이용정원=[]
심사여부=[]
이용요금=[]
미니홈페이지=[]

#상세설명
상세설명=[]
```

```python
driver = webdriver.Chrome('chromedriver')
driver.get('https://www.eshare.go.kr/UserPortal/adv/resources/AdvEsSearchResourcesMainView.do?rsrcClsCd=010500&sido=')
driver.implicitly_wait(3)
```

```python
select = Select(driver.find_element_by_xpath('//*[@id="dtlSido"]'))
select.select_by_visible_text('서울특별시')
driver.find_element_by_xpath('//*[@id="UprResrcSearchForm"]/article/div/div[2]/a').click()
```

```python
for page in range(16,16+1): #2~10페이지(1페이지는 디폴트) 
    for content in range(1,1+1):          
        driver.implicitly_wait(15)
        driver.find_element_by_xpath('//*[@id="searchList"]/div[1]/ul/li['+str(content)+']/div/div/div[2]/em').click()
        driver.implicitly_wait(10)
        try:    
            #요일별 이용시간
            평일.append(driver.find_element_by_xpath('//*[@id="detail_head"]/div[2]/div[2]/div/div/table/tbody/tr[1]/td').text.strip())
            토요일.append(driver.find_element_by_xpath('//*[@id="detail_head"]/div[2]/div[2]/div/div/table/tbody/tr[2]/td').text.strip())
            일요일_공휴일.append(driver.find_element_by_xpath('//*[@id="detail_head"]/div[2]/div[2]/div/div/table/tbody/tr[3]/td').text.strip())
            접수일시.append(driver.find_element_by_xpath('//*[@id="detail_head"]/div[2]/div[2]/div/div/table/tbody/tr[4]/td').text.strip())
            이용시간.append(driver.find_element_by_xpath('//*[@id="detail_head"]/div[2]/div[2]/div/div/table/tbody/tr[5]/td').text.strip())
            driver.implicitly_wait(10)
        except:
            driver.implicitly_wait(15)
            평일.append("")
            토요일.append("")
            일요일_공휴일.append("")
            접수일시.append("")
            이용시간.append("")
            driver.implicitly_wait(10)  
            
            #기본정보
        자원분류.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[1]/td[1]').text.strip())
        자원명칭.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[1]/td[2]').text.strip())
        장소_위치.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[2]/td[1]').text.strip())
        제공기관.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[2]/td[2]').text.strip())
        담당자정보.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[3]/td[1]').text.strip())
        예약방법.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[3]/td[2]').text.strip())
        예약문의.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[4]/td[1]').text.strip())
        이용대상.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[4]/td[2]').text.strip())
        준비시간.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[5]/td[1]').text.strip())
        신청서식.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[5]/td[2]').text.strip())
        이용정원.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[6]/td[1]').text.strip())
        심사여부.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[6]/td[2]').text.strip())
        이용요금.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[7]/td[1]').text.strip())
        미니홈페이지.append(driver.find_element_by_xpath('//*[@id="detail_summery"]/div[2]/div/div/table/tbody/tr[7]/td[2]').text.strip())
            #상세설명
        detail=driver.find_element_by_xpath('//*[@id="detail_descript"]/div').text.split("\n")
        상세설명.append(" ".join(detail))

        driver.back() #뒤로 가기
        driver.implicitly_wait(10)
        #######################################서울시 검색##########################################
        select = Select(driver.find_element_by_xpath('//*[@id="dtlSido"]')) 
        select.select_by_visible_text('서울특별시') 
        driver.find_element_by_xpath('//*[@id="UprResrcSearchForm"]/article/div/div[2]/a').send_keys(Keys.ENTER)
        time.sleep(4)
        #####################################11페이지로 넘어가기####################################
        driver.find_element_by_xpath('//*[@id="page-link"]').click()
        driver.implicitly_wait(20)  
        driver.find_element_by_xpath('//*[@id="page-link'+str(page)+'"]').click() #본인 페이지 넘김
        driver.implicitly_wait(10)
    driver.find_element_by_xpath('//*[@id="page-link'+str(page+1)+'"]').click() #1페이지 넘김
    driver.implicitly_wait(10)
```
